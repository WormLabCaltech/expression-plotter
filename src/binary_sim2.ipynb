```python
df = pd.read_csv('../simulation_output/binary/sim2.csv')
```

```python
df.head()
```

```python
sns.lmplot(data=df, x='dist', y='p', hue='n', ci=None, scatter_kws={'alpha':0.5}, line_kws={'alpha':0.7})
```

```python
df['log_p'] = -df.p.apply(np.log10)

sns.lmplot(x='dist', y='log_p', hue='n', data=df,
           ci=None, scatter_kws={'alpha':0.25}, line_kws={'alpha':0.8}, robust=True, size=7)
plt.axhline(-np.log10(0.05), ls='--', lw=1, color='k')
plt.axvline(.333, ls='--', lw=1, color='k')

plt.xlabel('Dist')
plt.ylabel('$-\log_{10}p$')
```
